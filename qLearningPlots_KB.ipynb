```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
columns = ['SubjID', 'Full_ID', 'TMS', 'Beta_All', 'Pers_All', 'Alpha_Pos_All', 'Alpha_Neg_All', 'Beta_Sixty', 'Pers_Sixty', 'Alpha_Pos_Sixty', 'Alpha_Neg_Sixty']
path = '/Users/maltegueth/Documents/qParams_opioid_AllData.txt'
f = pd.read_csv(path, sep='\t', header=1, names=columns)
xticklabels = ['Active', 'Sham']

f
```

```python
data = f[f['TMS']!=3]
data
```

# Histograms for all subjects

```python
font = {'family' : 'Sans',
        'weight' : 'normal',
        'size'   : 18}

matplotlib.rc('font', **font)
```

```python
fig1, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)

sns.despine(left=True)
sns.distplot(data['Beta_All'], kde=False, color="b", ax=axes[0,0])
sns.distplot(data['Beta_All'], hist=False, rug=True, color="r", ax=axes[0, 1])
sns.distplot(data['Beta_All'], hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])
sns.distplot(data['Beta_All'], color="m", ax=axes[1, 1])
plt.setp(axes, yticks=[])
plt.tight_layout()
```

```python
fig2, axes = plt.subplots(2, 4, figsize=(14, 8), sharex=True)
            
sns.distplot(data['Beta_All'], color="b", ax=axes[0,0])

sns.distplot(data['Pers_All'], color="b", ax=axes[0,1])

sns.distplot(data['Alpha_Pos_All'], color="b", ax=axes[1,1])

sns.distplot(data['Alpha_Neg_All'], color="b", ax=axes[1,0])

sns.distplot(data['Beta_Sixty'], color="b", ax=axes[0,2])

sns.distplot(data['Pers_Sixty'], color="b", ax=axes[0,3])

sns.distplot(data['Alpha_Neg_Sixty'], color="b", ax=axes[1,2])

sns.distplot(data['Alpha_Pos_Sixty'], color="b", ax=axes[1,3])

plt.setp(axes, xticks=np.arange(-1,3,1))

plt.tight_layout()
```

```python
fig3, axes = plt.subplots(2, 4, figsize=(14, 8), sharex=True)
            
sns.distplot(data['Beta_All'], hist=False, rug=True, color="b", ax=axes[0,0])

sns.distplot(data['Pers_All'], hist=False, rug=True, color="b", ax=axes[0,1])

sns.distplot(data['Alpha_Pos_All'], hist=False, rug=True, color="b", ax=axes[1,1])

sns.distplot(data['Alpha_Neg_All'], hist=False, rug=True, color="b", ax=axes[1,0])

sns.distplot(data['Beta_Sixty'], hist=False, rug=True, color="b", ax=axes[0,2])

sns.distplot(data['Pers_Sixty'], hist=False, rug=True, color="b", ax=axes[0,3])

sns.distplot(data['Alpha_Neg_Sixty'], hist=False, rug=True, color="b", ax=axes[1,2])

sns.distplot(data['Alpha_Pos_Sixty'], hist=False, rug=True, color="b", ax=axes[1,3])

plt.setp(axes, xticks=np.arange(-1,3,1))

plt.tight_layout()
```

```python
fig4, axes = plt.subplots(2, 4, figsize=(14, 8), sharex=True)

TMS = data[data['TMS'] == 1]
Sham = data[data['TMS'] == 2]

sns.distplot(TMS['Beta_All'], color="b", ax=axes[0,0])
sns.distplot(Sham['Beta_All'], color="r", ax=axes[0,0])

sns.distplot(TMS['Pers_All'], color="b", ax=axes[0,1])
sns.distplot(Sham['Pers_All'], color="r", ax=axes[0,1])

sns.distplot(TMS['Alpha_Pos_All'], color="b", ax=axes[1,1])
sns.distplot(Sham['Alpha_Pos_All'], color="r", ax=axes[1,1])

sns.distplot(TMS['Alpha_Neg_All'], color="b", ax=axes[1,0])
sns.distplot(Sham['Alpha_Neg_All'], color="r", ax=axes[1,0])

sns.distplot(TMS['Beta_Sixty'], color="b", ax=axes[0,2])
sns.distplot(Sham['Beta_Sixty'], color="r", ax=axes[0,2])

sns.distplot(TMS['Pers_Sixty'], color="b", ax=axes[0,3])
sns.distplot(Sham['Pers_Sixty'], color="r", ax=axes[0,3])

sns.distplot(TMS['Alpha_Neg_Sixty'], color="b", ax=axes[1,2])
sns.distplot(Sham['Alpha_Neg_Sixty'], color="r", ax=axes[1,2])

sns.distplot(TMS['Alpha_Pos_Sixty'], color="b", ax=axes[1,3])
sns.distplot(Sham['Alpha_Pos_Sixty'], color="r", ax=axes[1,3])

plt.setp(axes[0,0], yticks=np.arange(1,22,2))
plt.setp(axes[1,1], yticks=np.arange(1,20,2))
plt.setp(axes, xticks=np.arange(-1,3,1))
axes[0,0].legend(['TMS', 'Sham'], loc='upper right')

plt.tight_layout()
```

# Violinplots over all subjects

```python
fig6, ax1 = plt.subplots(2, 4, figsize=(14, 8), sharex=True)

sns.violinplot(y="Beta_All", data=data, palette="GnBu_d", saturation=1, ax=ax1[0,0])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Beta_All')
sns.despine(offset=10, trim=True)

sns.violinplot(y="Pers_All", data=data, palette="GnBu_d", saturation=1, ax=ax1[0,1])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Pers_All')
sns.despine(offset=10, trim=True)

sns.violinplot(y="Beta_Sixty", data=data, palette="GnBu_d", saturation=1, ax=ax1[0,2])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Beta_Sixty')
sns.despine(offset=10, trim=True)

sns.violinplot(y="Pers_Sixty", data=data, palette="GnBu_d", saturation=1, ax=ax1[0,3])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Pers_Sixty')
sns.despine(offset=10, trim=True)

sns.violinplot(y="Alpha_Pos_All", data=data, palette="GnBu_d", saturation=1, ax=ax1[1,0])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Alpha_Pos_All')
sns.despine(offset=10, trim=True)

sns.violinplot(y="Alpha_Neg_All", data=data, palette="GnBu_d", saturation=1, ax=ax1[1,1])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Alpha_Pos_All')
sns.despine(offset=10, trim=True)

sns.violinplot(y="Alpha_Pos_Sixty", data=data, palette="GnBu_d", saturation=1, ax=ax1[1,2])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Alpha_Pos_All')
sns.despine(offset=10, trim=True)

sns.violinplot(y="Alpha_Neg_Sixty", data=data, palette="GnBu_d", saturation=1, ax=ax1[1,3])
plt.setp(ax1, xticklabels=xticklabels)
plt.setp(ax1, ylabel='Alpha_Pos_All')
sns.despine(offset=10, trim=True)


plt.tight_layout()
```
